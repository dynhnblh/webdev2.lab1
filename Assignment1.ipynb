```python
ds.replace("?",np.nan,inplace=True)
ds
```

```python
missing_data = ds.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
ds['bore'] = ds['bore'].fillna(0)
ds['horsepower'] = ds['horsepower'].fillna(0)
ds['peak-rpm'] = ds['peak-rpm'].fillna(0)
```

```python
ds['num-of-doors'].value_counts()
ds['num-of-doors'].value_counts().idxmax()
ds["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
avg_norm_loss = ds["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
ds["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
avg_bore=ds['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
ds.dropna(subset=["price"], axis=0, inplace=True)
ds.reset_index(drop=True, inplace=True)
```

```python
ds[["bore", "stroke"]] = ds[["bore", "stroke"]].astype("float")
ds[["normalized-losses"]] = ds[["normalized-losses"]].astype("int")
ds[["price"]] = ds[["price"]].astype("float")
ds[["peak-rpm"]] = ds[["peak-rpm"]].astype("float")
```
